```python
import time
import random
import matplotlib.pyplot as plt

# Data nama yang sudah diurutkan
names_list = sorted([
    "Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah", "Isaac", "Jack",
    "Kara", "Liam", "Mona", "Nina", "Oliver", "Paul", "Quinn", "Rita", "Steve", "Tina",
    "Uma", "Vera", "Walter", "Xander", "Yara", "Zane", "Anna", "Ben", "Clara", "Derek",
    "Ella", "Felix", "Gina", "Hank", "Ivy", "James", "Kathy", "Leo", "Mason", "Nora", "Oscar",
    "Pam", "Quincy", "Ralph", "Sara", "Tom", "Ursula", "Vince", "Wendy", "Xena", "Yvonne",
    "Zoe", "Aaron", "Beatrice", "Clyde", "Danielle", "Eva", "Fiona", "Gareth", "Holly",
    "Ian", "Joan", "Kevin", "Lydia", "Mark", "Nancy", "Olga", "Peter", "Rebecca", "Simon",
    "Tracy", "Ulysses", "William", "Yasmin", "Aidan", "Brenna", "Cedric", "Daphne", "Emmett",
    "Felicity", "Gabriel", "Helena", "Irene", "Jackie", "Lennox", "Marvin", "Naomi", "Olivia",
    "Preston", "Rosie", "Sebastian", "Toby", "Violet", "Warren", "Zara", "Amos", "Bella",
    "Faye", "Grant", "Jordan", "Kirsten", "Lucas", "Maggie", "Penny", "Riley", "Sandra", "Theo",
    "Victor", "Willa", "Blake", "Chloe", "Dexter", "Ethan", "Flora", "Monica", "Nathan", "Olivia",
    "Sarah", "Theo", "Ursula", "Vince", "Annie", "Benny", "Cleo", "Dylan", "Eli", "Fiona", "George",
    "Harry", "Irene", "Jake", "Kris", "Lydia", "Maya", "Nick", "Olga", "Penny", "Rachel", "Samuel",
    "Tara", "Victor", "Willa", "Alex", "Beth", "Carl", "Diana", "Edward", "Fiona", "George", "Hannah",
    "Jacob", "Kylie", "Leo", "Mason", "Nina", "Oscar", "Paul", "Rita", "Steve", "Uma", "Vera", "Walter",
    "Yara", "Zane", "Amanda", "Barry", "Catherine", "Dominic", "Eliza", "Fay", "Garry", "Harriet",
    "Ingrid", "Julian", "Kim", "Lola", "Megan", "Nathaniel", "Olive", "Penelope", "Quincy", "Riley",
    "Sophia", "Toby", "Vicky", "Zane", "Alice", "Ben", "Catherine", "David", "Ella", "Felix", "Gina",
    "Harry", "Isabelle", "Jared", "Katrina", "Liam", "Mona", "Nina", "Olivia", "Peter", "Quincy",
    "Rebecca", "Sarah", "Theo", "Ursula", "Vince", "Wendy", "Xena", "Yvonne", "Zoe", "Annie", "Benny",
    "Cleo", "Dylan", "Eli", "Fiona", "George", "Harry", "Irene", "Jake", "Kris", "Lydia", "Maya", "Nick",
    "Olga", "Penny", "Rachel", "Samuel", "Tara", "Victor", "Willa", "Xander", "Yara", "Zara", "Alice",
    "Bob", "Carl", "Diane", "Evan", "Fiona", "Gail", "Henry", "Isla", "Jack", "Lola", "Maya", "Nina",
    "Oscar", "Paul", "Quincy", "Rita", "Sara", "Tom", "Uma", "Vera", "Walter", "Xander", "Yasmin", "Zane",
    "Melody", "Dustin", "Sylvia", "Olivera", "Gordon", "Tina", "Samuel", "Linda", "Cory", "Aubrey",
    "Freddie", "Whitney", "Lorna", "Martin", "Pablo", "Jordan", "Audrey", "Leona", "Shannon", "Kurt",
    "Rafael", "Carla", "Elsie", "Gina", "Preston", "Zara", "Rowan", "Julian", "Lucas", "Nadia", "Paige",
    "Dennis", "Billy", "Stella", "Rafael", "Tina", "Zoe", "Maggie", "Nico", "Holly", "Marina", "Archer",
    "Blaine", "Rosie", "Sandy", "Paola", "Callum", "Tobias", "Lyle", "Yasmin", "Quentin", "Derek", "Zane",
    "Wendy", "Olivia", "Naomi", "Daniela", "Zoe", "Emil", "Jack", "Ruben", "Ava", "Grace", "Eric", "Harrison",
    "Yvonne", "Margot", "Lilly", "Taylor", "Noah", "Brian", "Brianna", "Carmen", "Travis", "Emmanuel", "Iris",
    "Aiden", "Bella", "Caden", "Diana", "Eleanor", "Finn", "Gage", "Hailey", "Ian", "Jade",
    "Kai", "Lena", "Mason", "Nova", "Owen", "Piper", "Quinn", "Ryder", "Sage", "Tyler",
    "Vera", "Wade", "Xander", "Yara", "Zane", "Amelia", "Blake", "Chloe", "Dylan", "Eve",
    "Felix", "Grace", "Hannah", "Isla", "Jack", "Lily", "Mia", "Nathan", "Olivia", "Parker",
    "Quincy", "Riley", "Sophia", "Theo", "Uma", "Violet", "Will", "Xander", "Yasmine", "Zoe",
    "Ava", "Ben", "Caden", "Derek", "Eva", "Flora", "Gavin", "Holly", "Ian", "Jasper",
    "Kara", "Levi", "Maya", "Nina", "Ollie", "Penny", "Quinn", "Rose", "Sophie", "Toby",
    "Ursula", "Vince", "Willow", "Xander", "Yara", "Zane", "Ariana", "Brady", "Carla",
    "David", "Emma", "Finn", "Gina", "Hannah", "Ivy", "Jack", "Kendall", "Luca", "Mila",
    "Noah", "Olivia", "Paul", "Quincy", "Rita", "Sofia", "Tess", "Uma", "Vera", "Wendy",
    "Xander", "Yasmin", "Zara", "Adeline", "Blake", "Casey", "Dante", "Elliott", "Freya",
    "Gage", "Harper", "Ira", "Jasper", "Kara", "Lana", "Megan", "Nina", "Oscar", "Penny",
    "Quinn", "Rory", "Sierra", "Theo", "Ulysses", "Vince", "Willa", "Xander", "Yasmine",
    "Zoe", "Aiden", "Brady", "Carmen", "Dorian", "Emory", "Felix", "Gemma", "Haley", "Indigo",
    "Jack", "Kendall", "Luca", "Mason", "Nina", "Olivia", "Penny", "Quincy", "Riley", "Sophia",
    "Theo", "Una", "Vince", "Wendy", "Xander", "Yasmin", "Zane", "Amelia", "Bryce", "Clara",
    "Dylan", "Ella", "Fiona", "Gavin", "Holly", "Isla", "Jordan", "Keira", "Liam", "Megan",
    "Nora", "Oliver", "Penny", "Quinn", "Ryan", "Sophie", "Theo", "Vera", "Willow", "Xander",
    "Yasmin", "Zoe", "Ava", "Brady", "Caden", "Dylan", "Eli", "Freya", "Gracie", "Hugo", "Ivy",
    "Jack", "Kara", "Luca", "Mason", "Nina", "Olivia", "Penny", "Quinn", "Riley", "Sophie",
    "Theo", "Uma", "Vince", "Wendy", "Xander", "Yasmin", "Zara", "Amos", "Brock", "Carmen",
    "Diana", "Elena", "Fiona", "Gale", "Holly", "Ian", "Jack", "Kara", "Luna", "Mia", "Nia",
    "Oscar", "Piper", "Quinn", "Rose", "Sam", "Theo", "Uma", "Vera", "Wendy", "Xander",
    "Yasmin", "Zane", "Alina", "Curtis", "Gail", "Brock", "zude", "Zoana", "Zoya", "Zozaa"
])

# Fungsi untuk membuat nomor telepon acak
def generate_phone_number(name):
    # Mengambil 3 huruf pertama dari nama untuk bagian awalan
    prefix = ''.join([char.lower() for char in name[:3]])

    # Membuat nomor telepon acak dengan format +62 8X-XXXX-XXXX
    phone_number = f"+62 8{random.randint(10, 99)}-{random.randint(1000, 9999)}-{random.randint(1000, 9999)}"
    return phone_number

# Algoritma Binary Search
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Algoritma Linear Search
def linear_search_recursive(arr, target, index=0):
    if index >= len(arr):
        return -1
    if arr[index] == target:
        return index
    return linear_search_recursive(arr, target, index + 1)

# Membuat direktori untuk nomor telepon
phone_book = {name: generate_phone_number(name) for name in names_list}

# Pilih posisi target: awal, tengah, dan akhir
positions = [0, len(names_list) // 2, len(names_list) - 1]
position_labels = ["Awal", "Tengah", "Akhir"]

# Pengujian waktu untuk setiap posisi target
results_binary = []
results_linear = []

for pos in positions:
    search_target = names_list[pos]

    # Waktu Binary Search
    start_binary = time.time()
    result_binary = binary_search(names_list, search_target)
    end_binary = time.time()

    # Waktu Linear Search
    start_linear = time.time()
    result_linear = linear_search_recursive(names_list, search_target)
    end_linear = time.time()

    # Menyimpan hasil waktu untuk visualisasi
    results_binary.append(end_binary - start_binary)
    results_linear.append(end_linear - start_linear)

    # Output
    print(f"Target: {search_target}")
    print(f"Phone Number: {phone_book[search_target]}")
    print(f"Binary Search Result: {names_list[result_binary] if result_binary != -1 else 'Not found'}, Data Index: {result_binary if result_binary != -1 else 'N/A'}, Time: {end_binary - start_binary:.10f} seconds")
    print(f"Linear Search Result: {names_list[result_linear] if result_linear != -1 else 'Not found'}, Data Index: {result_linear if result_linear != -1 else 'N/A'}, Time: {end_linear - start_linear:.10f} seconds")
    print('-' * 100)
```
